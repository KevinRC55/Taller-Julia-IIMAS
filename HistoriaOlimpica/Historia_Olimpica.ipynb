```julia
using CSV
using HTTP
using DataFrames
using StatsBase
using Plots

ENV["COLUMNS"] = 1000
```

```julia
url = "https://raw.githubusercontent.com/KevinRC55/Taller-Julia-IIMAS/master/HistoriaOlimpica/dictionary.csv"
paises = CSV.read(HTTP.get(url).body, DataFrame)
rename!(paises,:Country => :Country_Name)
rename!(paises,:Code => :Country)
```

```julia
url = "https://raw.githubusercontent.com/KevinRC55/Taller-Julia-IIMAS/master/HistoriaOlimpica/summer.csv"
juegos = CSV.read(HTTP.get(url).body, DataFrame)
```

```julia
describe(paises, cols= :, :nmissing)
```

```julia
describe(juegos, cols= :, :nmissing)
```

```julia
dropmissing!(juegos, :Country)
juegos = leftjoin(juegos, paises, on = :Country)
select!(juegos, 1:3, :Country, :Country_Name, 4:5, 7:9, Not([:Population, Symbol("GDP per Capita")]))
```

```julia
first(juegos, 30)
```

```julia
last(juegos, 30)
```

## Sedes Olímpicas

```julia
cty = countmap(juegos.City)
x = collect(keys(cty))
y = collect(values(cty))
pie(x, y, title="Sedes Olimpicas", l=0.5, shadow=true,)
```

## Atletas

```julia
Plots.plotly() 
gndr = countmap(juegos.Gender)
x = collect(keys(gndr))
y = collect(values(gndr))
pie(x, y, title="Medallas por género", l=0.5, shadow=true)
```

### Atletas con más medallas

```julia
atletas = sort(combine(groupby(juegos, [:Athlete, :Country, :Sport]), nrow => :Medals), :Medals, rev = true)
```

### Atletas con más medallas por país

```julia
filter(row -> row.Country ∈ ["MEX"], atletas)
```

### Atletas con más medallas por deporte

```julia
filter(row -> row.Sport ∈ ["Taekwondo"], atletas)
```

### Atletas con más medallas por año

```julia
year = filter(row -> row.Year ∈ [1968], juegos)
year = sort(combine(groupby(year, [:Athlete, :Country, :Sport]), nrow => :Medals), :Medals, rev = true)
```

#### Atletas con más medallas de oro

```julia
golds = filter(row -> row.Medal ∈ ["Gold"], juegos)
golds = sort(combine(groupby(golds, [:Athlete, :Country, :Sport]), nrow => :Golds), :Golds, rev = true)
```

#### Atletas con más medallas de plata

```julia
silvers = filter(row -> row.Medal ∈ ["Silver"], juegos)
silvers = sort(combine(groupby(silvers, [:Athlete, :Country, :Sport]), nrow => :Silvers), :Silvers, rev = true)
```

#### Atletas con más medallas de bronce

```julia
bronzes = filter(row -> row.Medal ∈ ["Bronze"], juegos)
bronzes = sort(combine(groupby(bronzes, [:Athlete, :Country, :Sport]), nrow => :Bronzes), :Bronzes, rev = true)
```
